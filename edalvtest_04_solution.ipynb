```python
# 1-3
def get_country_code_2(country_name):
    import pycountry

    try:
        country = pycountry.countries.get(name=country_name)
        return country.alpha_2
    except:
        country = pycountry.countries.search_fuzzy(country_name)
        return country[0].alpha_2

df_target['code'] = df_target['country'].apply(get_country_code_2)
check_01_03(df_target)
```

```python
# 4-1
def weighted_avg(row):
    import numpy as np

    weights = row['population']
    target_internet = row['internet_use_rate']
    target_income = row['income_per_person']

    datas = {
        'weighted_ave_internet': np.average(target_internet, weights=weights),
        'weighted_ave_income': np.average(target_income, weights=weights),
        }

    return pd.Series(datas)

df_result = df.groupby(['region', 'sub_region']).apply(weighted_avg)
check_04_01(df_result)
```
