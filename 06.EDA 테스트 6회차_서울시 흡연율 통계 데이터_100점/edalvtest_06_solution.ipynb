````python
# 2-1

# 방법1
def get_female_ratio(row):
    import sympy

    if pd.notna(row.여자):
        x = sympy.symbols('x')
        f = sympy.Eq(row.여자 * x/100 + row.남자 * (1 - x/100), row.전체)
        f_ratio = round(float(sympy.solve(f)[0]), 6)
        return f_ratio
    else:
        return None
df_target_check = df_target.copy()
df_target_check['여자비율'] = df_target_check.apply(get_female_ratio, axis=1)

print('방법1')
check_02_01(df_target_check)
display(df_target_check.head())

def get_female_ratio(row):
    if pd.notna(row.여자):
        f_ratio = (row.전체 - row.남자) / (row.여자 - row.남자) * 100
        return round(f_ratio, 6)
    else:
        return None

df_target['여자비율'] = df_target.apply(get_female_ratio, axis=1)

print('방법2')
check_02_01(df_target)
display(df_target.head())
````
