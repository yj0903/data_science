```python
# 3-1

# 해당 문제에서는 데이터기준일자의 형식(년-월-일) 특성상 타입을 변경하지 않아도 정상적으로 대소비교 가능합니다..
# 따라서 아래 주석과 같이 format을 변경할 필요 없습니다.
# from datetime import date
# df_target.데이터기준일자 = df_target.데이터기준일자.apply(date.fromisoformat)

# 방법1
# 결과 dataframe에 index name이 지정되어있으나 방법2와 동일한 dataframe입니다.
df_target_temp = df_target.copy()

def replace_space(name):
    while ' ' in name:
        name = name.replace(' ', '')
    return name.strip()

df_target_temp['name_del_space'] = df_target_temp.시설명.apply(replace_space)

df_target_temp.drop(df_target_temp[df_target_temp.시설명.str.contains('휴관')].index, inplace=True)

not_unique_names = [name for name in df_target_temp.name_del_space.unique() if (df_target_temp.name_del_space == name).count() != 1]

df_target_temp.index.name = 'index'
df_target_temp.sort_values(by=['데이터기준일자', 'index'], ascending=[False, True], inplace=True)
df_target_temp = df_target_temp.loc[df_target_temp.name_del_space.drop_duplicates(keep='first').index]

df_target_temp.drop('name_del_space', axis=1, inplace=True)
df_target_temp.sort_index(inplace=True)
print('방법1')
check_03_01(df_target_temp)
display(df_target_temp.head())

# 방법2
df_target_temp = df_target.copy()

def replace_space(name):
    while ' ' in name:
        name = name.replace(' ', '')
    return name.strip()

df_target_temp['name_del_space'] = df_target_temp.시설명.apply(replace_space)

df_target_temp.drop(df_target_temp[df_target_temp.시설명.str.contains('휴관')].index, inplace=True)

not_unique_names = [name for name in df_target_temp.name_del_space.unique() if (df_target_temp.name_del_space == name).count() != 1]
for name in not_unique_names:
    temp_df = df_target_temp[(df_target_temp.name_del_space == name)]
    
    keep_idx = sorted(temp_df[(temp_df.데이터기준일자 == temp_df.데이터기준일자.max())].index)[0]
    del_idx = temp_df.index.drop(keep_idx)
    
    df_target_temp.drop(del_idx, inplace=True)
    
df_target_temp.drop('name_del_space', axis=1, inplace=True)
print('방법2')
check_03_01(df_target_temp)
display(df_target_temp.head())

df_target = df_target_temp.copy()
```

```python
# 3-2

# 아래 방법들은 새벽에 관람을 종료하는 박물관/미술관이 없음을 확인했기 때문에 가능한 코드입니다. 
# 만약 10:00(관람시작시각) ~ 02:00(관람종료시각)이라면, 하기 코드로는 음수(-8)가 나오기때문에 24를 더한 16시간이라고 표기되어야 합니다.

# 방법1
df_target_temp = df_target.copy()
def get_open_hour(row):
    from datetime import datetime
    
    now = datetime.now()

    open_str = row[row.index.str.contains('시작시각')][0]
    hour, minute = map(int, open_str.split(':'))
    open_datetime = now.replace(hour=hour, minute=minute, second=0, microsecond=0)

    close_str = row[row.index.str.contains('종료시각')][0]
    hour, minute = map(int, close_str.split(':'))
    close_datetime = now.replace(hour=hour, minute=minute, second=0, microsecond=0)

    open_time = round((close_datetime - open_datetime).seconds / 3600, 2)
    open_time = 24 if open_time > 23 else open_time
        
    return open_time

days_list = ['평일', '공휴일']
for day_str in days_list:
    target_cols = [f'{day_str}관람시작시각', f'{day_str}관람종료시각']
    df_target_temp[f'{day_str}관람가능시간'] = df_target_temp[target_cols].apply(get_open_hour, axis=1)
print('방법1')
check_03_02(df_target_temp)
display(df_target_temp.head())

# 방법2
df_target_temp = df_target.copy()
def get_open_hour(row):

    open_str = row[row.index.str.contains('시작시각')][0]
    hour, minute = map(int, open_str.split(':'))
    open_float = sum([hour, minute/60])

    close_str = row[row.index.str.contains('종료시각')][0]
    hour, minute = map(int, close_str.split(':'))
    close_float = sum([hour, minute/60])

    open_time = round(close_float - open_float, 2)
    open_time = 24 if open_time > 23 else open_time
        
    return open_time

days_list = ['평일', '공휴일']
for day_str in days_list:
    target_cols = [f'{day_str}관람시작시각', f'{day_str}관람종료시각']
    df_target_temp[f'{day_str}관람가능시간'] = df_target_temp[target_cols].apply(get_open_hour, axis=1)
print('방법2')
check_03_02(df_target_temp)
display(df_target_temp.head())

df_target = df_target_temp.copy()
```
